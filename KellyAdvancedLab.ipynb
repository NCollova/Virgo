## Tranforming to Supergalactic Coordinates##

Looking to match the plots shown in Kim+2016

```python
#Plot of Virgo Cluster in galactic coordinates
sc = SkyCoord(NSA.RA_1*u.degree, NSA.DEC_1*u.degree)
galcoords = sc.supergalactic

l = galcoords.sgl.degree
b = galcoords.sgb.degree

distance_z = NSA.ZDIST
SGX = distance_z*c.to('km/s')/cosmo.H(0)*np.cos(l * np.pi / 180.)*np.cos(b * np.pi / 180.)
SGY = distance_z*c.to('km/s')/cosmo.H(0)*np.sin(l * np.pi / 180.)*np.cos(b * np.pi / 180.)
SGZ = distance_z*c.to('km/s')/cosmo.H(0)*np.sin(b * np.pi / 180.)


l = 102.93 # longitude, deg
b = -2.73 # latitude, deg

ckms= c.to('km/s')
SGX_Virgo = distance_z*c.to('km/s')/cosmo.H(0)*np.cos(l * np.pi / 180.)*np.cos(b * np.pi / 180.)
SGY_Virgo = distance_z*c.to('km/s')/cosmo.H(0)*np.sin(l * np.pi / 180.)*np.cos(b * np.pi / 180.)
SGZ_Virgo = distance_z*c.to('km/s')/cosmo.H(0)*np.sin(b * np.pi / 180.)

# difference 
DSGX = SGX - SGX_Virgo
DSGY = SGY - SGY_Virgo
DSGZ = SGZ - SGZ_Virgo

# don't use objects w/in 3.6 Mpc of Virgo (Kim+2016)

dist = np.sqrt(DSGX**2 + DSGY**2 + DSGZ**2) 
dist_flag = dist.value > 3.6
```

```python
kim_vflag = NSA.Z*3.e5 < 6000.  
kim_raflag = (NSA.RA_1 > 115.) & (NSA.RA_1 < 240.)
kim_decflag = (NSA.DEC_1 > -35.) & (NSA.DEC_1 < 60.)
kim_flag = kim_vflag & kim_raflag & kim_decflag & dist_flag


#In front of cluster
SGYfront_flag = (SGY > 4. * u.Mpc) & (SGY < 16. * u.Mpc)
SGYback_flag = (SGY > 16. * u.Mpc) & (SGY < 27. * u.Mpc)
```
